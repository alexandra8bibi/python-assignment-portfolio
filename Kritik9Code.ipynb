```python
# Part A
from sympy import symbols, exp, sin, diff
# Importing exp and sin directly for easier access

# Define symbols
x, y = symbols('x y')

# Define the function
function = exp(x) * sin(y) + y**3

# Calculate partial derivatives
partial_derivative_x = diff(function, x)
partial_derivative_y = diff(function, y)

# Display the partial derivatives
print(f"Partial derivative with respect to x: {partial_derivative_x}")
print(f"Partial derivative with respect to y: {partial_derivative_y}")
```

```python
import sympy as sym

# Defining symbols
x, y = sym.symbols('x y')

# Function definition
f = (x**3)*2*y + (x*(y**2))

# Partial derivatives
partial_x = sym.diff(f, x)
partial_y = sym.diff(f, y)

# Gradient as a tuple
gradient = (partial_x, partial_y)

# Coordinates for evaluation
x_point = 1
y_point = -1

# Functions to evaluate the gradient components at a point
def gradient_x_at_point(x_val, y_val):
    return 2 * x_val * y_val + y_val**2

def gradient_y_at_point(x_val, y_val):
    return x_val**2 * 2 + 2 * x_val * y_val

# Function to calculate the magnitude of the gradient at a point
def magnitude_at_point(x_val, y_val):
    grad_x = gradient_x_at_point(x_val, y_val)
    grad_y = gradient_y_at_point(x_val, y_val)
    return sym.sqrt(grad_x**2 + grad_y**2)

# Evaluating the gradient and magnitude at the point (1, -1)
gradient_at_point = (gradient_x_at_point(x_point, y_point), gradient_y_at_point(x_point, y_point))
magnitude = magnitude_at_point(x_point, y_point)

# Printing the results
print(f"Gradient at ({x_point}, {y_point}) is: {gradient_at_point}")
print(f"Magnitude at ({x_point}, {y_point}) is: {magnitude}")
```

```python
import sympy as sp

# Define the variables
x, y = sp.symbols('x y')

# Define the function
func = sp.log(x**2 + 2*y**2)

# Compute second partial derivatives
second_derivative_x = sp.diff(sp.diff(func, x), x)
second_derivative_y = sp.diff(sp.diff(func, y), y)
mixed_derivative = sp.diff(sp.diff(func, x), y)  # This computes the mixed partial derivative

# Print the results
print(f"Second partial derivative with respect to x: {second_derivative_x}")
print(f"Second partial derivative with respect to y: {second_derivative_y}")
print(f"Mixed partial derivative: {mixed_derivative}")

# Comment on symmetry
# According to Clairaut's Theorem, the mixed partial derivatives fxy and fyx are equal if they are continuous.
if second_derivative_x == second_derivative_y:
    print("As per Clairaut's Theorem, the mixed partial derivatives are symmetrical.")
```

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
x_symbol, y_symbol = sym.symbols('x y')

# Define the expression
expression = x_symbol**2 - 2*x_symbol*y_symbol + y_symbol**2

# Convert the sympy expression to a lambda function for numerical evaluation
expression_func = sym.lambdify((x_symbol, y_symbol), expression, 'numpy')

# Generate values for x and y
x_values = np.linspace(-3, 3, 400)
y_values = np.linspace(-3, 3, 400)

# Create a meshgrid which will be used for plotting
X, Y = np.meshgrid(x_values, y_values)

# Evaluate the expression on the meshgrid
Z = expression_func(X, Y)

# Plot the contour map with more levels for smoother gradients
plt.contourf(X, Y, Z, levels=100, cmap='plasma')  # Changed colormap to 'plasma' for a different look
plt.colorbar()  # Add a colorbar to the plot

# Set the title and axis labels with LaTeX formatting
plt.title(r'Contour Plot of $f(x, y) = x^2 - 2xy + y^2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Show the plot
plt.show()
```
