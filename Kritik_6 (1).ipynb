```python
##PART 1 (defining the function)
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

def normal_density():
    variance = float(input("Input the σ^2 (variance) value here: "))
    mean = float(input("Input the μ (mean) value here: "))
    x = float(input("Input your x value here: "))

    normal_density = 1 / (sqrt(2 * pi * variance)) * np.exp(-(x - mean)**2 / (2 * variance))

    return normal_density

result = normal_density()
print("Normal Density:", result)
```

```python
##PART 2 (plotting)
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

def normal_density(x, mean, variance):
    return 1 / (sqrt(2 * pi * variance)) * np.exp(-(x - mean)**2 / (2 * variance))

# Generate x values
x_values = np.linspace(-10, 10, 1000)

# random different mean and variance values
mean_values = [0, 2, -1, 3, 5]
variance_values = [1, 0.5, 2, 6, 5]

for mean, variance in zip(mean_values, variance_values):
    y_values = normal_density(x_values, mean, variance)
    label = f'Mean={mean}, Variance={variance}'
    plt.plot(x_values, y_values, label=label)

plt.title('Normal Density Function for Different Mean and Variance')
plt.xlabel('x')
plt.ylabel('Normal Density')
plt.legend()
plt.show()
```

```python
##PART 3 (numerical integration)
import numpy as np
from scipy.integrate import quad
from math import sqrt, pi, exp

def integration_of_normal_density(mean, variance, a, b):
    result, _ = quad(lambda x: normal_density(x, mean, variance), a, b)
    return result

# Example
mean_value = 0 # could also change code to ask for user input if desired
variance_value = 1 # could also change code to ask for user input if desired
lower_bound = -1 # a value
upper_bound = 1 # b value

integral_result = integrate_normal_density(mean_value, variance_value, lower_bound, upper_bound)
print(f"Numerical Integration Result: {integral_result}")
```
